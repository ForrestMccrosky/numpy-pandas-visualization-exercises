## Exercise Number 5
    Using one chart to graph several equations

```python
import matplotlib.pyplot as plt
import math
```

```python
x = range(1, 21)
y1 = [0 * n + 1 for n in x]
y2 = [math.log(n) for n in x]
y3 = [n for n in x]
y4 = [n * math.log(n) for n in x]
y5 = [n ** 2 for n in x]
y6 = [2 ** n for n in x]
y7 = [math.factorial(n) for n in x]
y8 = [n ** n for n in x]

plt.figure(figsize = (12,8))
plt.title('Big O Notation', fontsize = 16)
plt.xlabel('Elements', fontsize = 14)
plt.ylabel('Operations', fontsize = 14)
plt.plot(x, y1, c = 'r', label = '$O(1)$')
plt.plot(x, y2, c = 'g', label = '$O(log n)$')
plt.plot(x, y3, c = 'b', label = '$O(n)$')
plt.plot(x, y4, c = 'c', label = '$O(n log n)$')
plt.plot(x, y5, c = 'y', label = '$O(n^2)$')
plt.plot(x, y6, c = 'm', label = '$O(2^n)$')
plt.plot(x, y7, c = 'k', label = '$O(n!)$')
plt.plot(x, y8, c = 'slategray', label = '$O(n^n)$')
plt.legend()



plt.ylim(0, 100)
plt.show()
```
