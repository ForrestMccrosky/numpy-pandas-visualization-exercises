## Exercises Part II Advanced Dataframes

```python
users = pd.DataFrame({
    'id': [1, 2, 3, 4, 5, 6],
    'name': ['bob', 'joe', 'sally', 'adam', 'jane', 'mike'],
    'role_id': [1, 2, 3, 3, np.nan, np.nan]
})
users
```

```python
roles = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'name': ['admin', 'author', 'reviewer', 'commenter']
})
roles
```

Question 2: What is the result of using a right join on the DataFrames?

```python
users.merge(roles, left_on='role_id', right_on='id', how='right', indicator=True)
```

Question 3: What is the result of using an outer join on the DataFrames?

```python
users.merge(roles, left_on='role_id', right_on='id', how='outer', indicator=True)
```

Question 4: What happens if you drop the foreign keys from the DataFrames and try to merge them?



```python
users.merge(roles, how='outer', indicator=True)
```

Questions 12 & 13: Create mileage_difference columns and average_mileage columns for the vehicles

```python
mpg['average_mileage2'] = mpg[['city','highway']].agg('mean', axis = 1)
mpg
```

```python
mpg['mileage_difference'] = mpg.highway - mpg.city
mpg['average_mileage'] = (mpg.city + mpg.highway) / 2
mpg
```

Question 14: Create a column named is_automatic determining if they are automatic or not

```python
mpg['is_automatic'] = np.where(mpg.trans.str.count('auto'), 'automatic', 'manual')
mpg
```

```python
mpg['is_automatic'] = np.where(mpg.trans.str.count('auto'), True, False)
mpg
```

Question 15: Using the mpg dataset, find out which manufacturer has the best miles per gallon on average?

```python
mpg.groupby('manufacturer').average_mileage.agg('mean').sort_values(ascending=False)
```

Question 16: Do automatic or manual cars have better miles per gallon?

```python
mpg.groupby('is_automatic').average_mileage.agg('mean').sort_values() #Looks like manual cars are the winners here
```
