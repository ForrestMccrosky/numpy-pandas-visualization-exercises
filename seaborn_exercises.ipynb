Question 1: What does the distribution of petal length look like

```python
sns.boxplot(data = iris, x='petal_length', y='species') ##boxplots are a good way to see the range, median, and outliers of the data
## easier to see the different distributions per species by setting the y value to species as well.
```

Question 2: Is there a correlation between petal length and petal width?

```python
sns.heatmap(iris.corr(), cmap = 'RdBu_r', 
            annot = True, vmax = 1, vmin = -1,
           linewidth = 1, linecolor = 'k')
## By using a heatmap to map the correlation of the iris dataframe we made
## we can see the box at petal_length and petal_width is a dark blue
## meaning there is correlation that says the petal_length increases 
## positively with petal_width
```

```python
sns.scatterplot(data = iris, x = 'petal_length', 
                y = 'petal_width', hue = 'species')
# more graphs to also view the different distributions of relationship
# between petal_length and petal_width
```

```python
sns.lmplot(x= 'petal_length', y = 'petal_width', 
           col = 'species', data = iris, truncate = False)
```

Question 3: would it be reasonable to predict species based on sepal width and sepal length?

```python
sns.relplot(data = iris, x = 'sepal_width', y = 'sepal_length',
            hue = 'species' 
           )
## we can see that it would easy to predict the species of setosa using 
## sepal_width and sepal_length. However, the distribution for the other 
## two species (versicolor and virginica) are two close to get an 
## completely accurate
## prediction using these columns
```

Question 4: Which features would be used to predict species?

```python
sns.scatterplot(data = iris, x = 'petal_length', 
                y = 'petal_width', hue = 'species')
## by scatter plotting the different flowers using petal width and length
## The distribution per species is almost perfectly clustered
## Making this the most accurate way to predict species of the flowers
```

Question 4: Using the chipotle dataset from the previous exercise, create a bar chart that shows the 4 most popular items and the revenue produced by each.

```python
def get_db_url(host,user,password):
    datatable = input('Please enter the database you wish to connect to: ')
    url = f'mysql+pymysql://{user}:{password}@{host}/{datatable}'
    return url
```

```python
url = get_db_url(host,user,password)
chipotle_df = pd.read_sql('Select * From orders', url)
chipotle_df
```

```python
chipotle_df['item_price'] = chipotle_df['item_price'].str.replace('$','').astype(float)
chipotle_df
```

```python
chipotle_df.item_name.value_counts().head(4)
```

```python
total_revenue = chipotle_df.groupby('item_name').sum('item_price').sort_values(by = 'item_price', ascending = False)
total_revenue
```

```python
top_four_items = total_revenue.sort_values(by = 'quantity', ascending = False).head(4)
```

```python
top_four_items
```

```python
sns.barplot(data = top_four_items, 
            y =  ['Chicken Bowl', 'Chicken Burrito', 'Chips and Guacamole', 'Steak Burrito'], 
            x = 'item_price')
```

Question 5: Load the sleepstudy data and read it's documentation. Use seaborn to create a line chart of all the individual subject's reaction times and a more prominant line showing the average change in reaction time.



```python
sleepstudy = data('sleepstudy')
sleepstudy
```

```python
data('sleepstudy', show_doc = True)
```

```python
plt.figure(figsize = (20,10))
sns.lineplot(data = sleepstudy, x = 'Days', y = 'Reaction', 
            hue = 'Subject', palette = 'rainbow',
           linewidth = 3, legend = 'full')
sns.lineplot(data = sleepstudy, x = 'Days', y = 'Reaction', 
            estimator = 'mean', linewidth = 10, ci = False)
plt.title('Reaction Times Vs. Sleep Deprivation', fontsize = 24)
plt.xlabel('Progression of Days', fontsize =18)
plt.ylabel('Recorded Reaction Times', fontsize = 18)
plt.text(1,400, 'Darker Orange Line = Average Reaction Time Change', fontsize = 14, color = 'r')
plt.show()
```
