## Matplotlib Exercises

Exercise 1: Use matplotlib to plot the following equation:
        
        y = x^2 - x + 2
        

```python
import matplotlib.pyplot as plt
```

```python
x = list(range(10))
y = [n ** 2 - n + 2 for n in x]
plt.plot(x,y)
plt.text(0, 5, '(0,0)', fontsize = 10, color = 'r')
```

```python
import math

x = list(range(10))
y1 = [math.sqrt(n) for n in x ]
plt.title('y = $\sqrt{x}$')
plt.plot(x,y1)
```

```python
x = list(range(10))
y2 = [n ** 3 for n in x ]
plt.title('y = $x^3$')
plt.plot(x,y2)
```

```python
x = list(range(10))
y3 = [2 ** n for n in x ]
plt.title('y = $2^x$')
plt.plot(x,y3)
```

```python
x = list(range(10))
y4 = [1/(n + 1) for n in x ]
plt.title('y = $1/(x+1)$')
plt.plot(x,y4)
```

## Use the 4 charts above to create one display with all 4
    # Using a subplot

```python
x = list(range(10))
plt.figure(figsize=(12,12))

plt.subplot(2,2,1)
plt.plot(x,y1)
plt.title('y = $\sqrt{x}$')
plt.xlabel('x: 1-10')
plt.ylabel('y values')


plt.subplot(2,2,2)
plt.plot(x,y2)
plt.title('y = $x^3$')
plt.xlabel('x: 1-10')
plt.ylabel('y values')


plt.subplot(2,2,3)
plt.plot(x,y3)
plt.title('y = $2^x$')
plt.xlabel('x: 1-10')
plt.ylabel('y values')


plt.subplot(2,2,4)
plt.plot(x,y4)
plt.title('y = $1/(x+1)$')
plt.xlabel('x: 1-10')
plt.ylabel('y values')


plt.suptitle('Different Equations')
plt.show()
```

## Now to combine all equations into one chart
    # Using one graph

```python

x = list(range(30))
plt.figure(figsize=(10,8))

y1 = [math.sqrt(n) for n in x ]
y2 = [n ** 3 for n in x ]
y3 = [2 ** n for n in x ]
y4 = [1/(n + 1) for n in x ]

plt.plot(x,y1, color ='red')
plt.plot(x,y2, color ='purple')
plt.plot(x,y3, color ='green')
plt.plot(x,y4, color ='blue')
plt.title("The Big Four", fontsize = 18)
plt.xlabel("The X factor's", fontsize = 12)
plt.ylabel("The Y factor's", fontsize = 12)
plt.plot(x, y1, color ='red', label = 'y = \sqrt(x)')
plt.plot(x, y2, color ='purple', label = 'y = $x^3$')
plt.plot(x, y3, color ='green', label = 'y = $2^x$')
plt.plot(x, y4, color ='blue', label = 'y = $1/(x+1)$')
plt.legend()
plt.ylim(0,200)
plt.show()
```
